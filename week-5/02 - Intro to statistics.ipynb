## Histograms

One way to understand the data is to count how many times each value appears in the sample.

```python
sample = [1,2,3,4,5,6]
```

```python
# Experiment: roll a die 20 times
exp = random.choices(sample, k=20)
exp
```

```python
# We make a comp dict
count_ = {n: exp.count(n) for n in exp}
count_
```

```python
# built-in alternative
from collections import Counter
```

```python
#Counter does the very same thing
Counter(count_)
```

```python
for k, v in Counter(exp).items():
    print(f"{k} -----> {v}")
```

And we could plot the result as a bar graph to have a visual representation.

```python
plt.bar(x=count_.keys(),
        height=count_.values())
plt.xlabel('Number on the die')
plt.ylabel('Frequency')
plt.show()
```

```python
# seaborn has a nice function to do this too that we already know very well
sns.countplot(x=exp, color='b');
```

We note that the above method is only valid for [discrete variables](http://onlinestatbook.com/glossary/discrete_variables.html). To make it more general, we can "split" the x-axis into pieces, so that observations within a certain group are counted together. In this way, the method described above will also work for [continuous variables](http://onlinestatbook.com/glossary/continuous_variables.html).

The pandas library has a `pd.cut` method that serves exactly that purpose.

```python
binned = pd.cut(exp, bins=3)
print(binned.value_counts())
```

```python
ax = binned.value_counts().plot.bar(rot=75, color="b", figsize=(6,4))
plt.show()
```

```python
np.histogram(exp, bins=6)
```

```python
plt.hist(exp, bins=4)
```

This that we have seen, also does numpy:

```python
np.histogram(exp, bins=6)
```

The above tables are called "frequency tables". A visual representation of a frequency table is what we call a [histogram](https://en.wikipedia.org/wiki/Histogram)

```python
plt.hist(exp, bins="auto")
plt.show
```

```python
# The difference with density is that it paints the probability distribution for me by putting density
plt.hist(exp, bins="auto", density=True)# Choosing the number of bins is misleading
# Density changes the numerical value for the percentage
plt.show()
```

*If an integer is given, the bin borders + 1 are calculated and returned, consistent with `numpy.histogram.`* All but the last (rightmost) bin is half open. In other words, if bins is `[1, 2, 3, 4]`
so the first bin is `[1, 2)` (including 1, but excluding 2) and the second `[2, 3)`.
However, the last bin is `[3, 4]`, which includes the 4

$$\bar{x} = \frac{\sum x_i}{N} = \frac{x_1 + x_2 + \dots + x_n}{N}$$

#### Truncated Mean

[Wikipedia:](https://en.wikipedia.org/wiki/Truncated_mean) A truncated mean or trimmed mean is a statistical measure of central tendency, much like the mean and median. It involves calculating the mean after discarding certain parts of a probability distribution or sample at the high and low ends, and usually discarding an equal number of both.

**Truncated mean**

```python
trim_mean(mock_salaries, 0.01)
```

```python
round(trim_mean(mock_salaries, 0.01), 10)
```

#### Median

[Wikipedia](https://en.wikipedia.org/wiki/Median): In statistics and probability theory, the median is a value that separates the upper half from the lower half of a data sample, a population, or a probability distribution. For a data set, it can be thought of as the "middle" value. The basic advantage of the median in describing data over the mean (often simply described as the "average") is that it is not biased as much by a small proportion of extremely large or extremely small values, and therefore can give a better idea of ​​a "typical" value.

```python
print(f'Median salary without Messi: {np.round(np.median(mock_salaries), 2)}')
print(f'Median salary with Messi: {np.round(np.median(salaries_with_messi), 2)}')
```

####  Mode

[Wikipedia:](https://en.wikipedia.org/wiki/Mode_(statistics)) The mode is the value that appears most often in a set of data values. Distributions can have more than one mode (or local maxima). These distributions are called *multimodal*.

In the case of discrete variables (or *quasi* discrete) it is a more or less easy quantity to calculate:

```python
df_titanic.age.mode()
```

But, when the variable is continuous, it most likely won't even have repeating values.
In that case you should fit your data to a known distribution (or an estimate) and calculate the maximum of the pdf. (See an example [here](https://stackoverflow.com/questions/59662920/how-to-get-the-mode-of-distribution-in-scipy-stats)).

```python
pd.Series(mock_salaries).plot.kde();
```

```python
kernel = gaussian_kde(mock_salaries)
height = kernel.pdf(mock_salaries)
mode_value = mock_salaries[np.argmax(height)]
print('Mode without Messi: ', mode_value)
```

```python
kernel = gaussian_kde(salaries_with_messi)
height = kernel.pdf(salaries_with_messi)
mode_value = mock_salaries[np.argmax(height)]
print('Mode with Messi: ', mode_value)
```

### Variance and standard deviation

In addition to knowing where a distribution is *centered*, it is also important to know how *variable* the data is, that is, how *spread* it is. Two statistics, the variance and the standard deviation, are useful in this case:

$$Var = \sigma^2 = \frac{\sum (x_i - \mu) ^ 2}{N}$$

$$Std = \sigma = \sqrt{Var}$$

From [Wikipedia:](https://en.wikipedia.org/wiki/Variance) In probability theory and statistics, variance is the expected squared deviation of a random variable from its mean. Informally, it measures the distance of a set of numbers from their mean value.

(**Note**: The above expressions refer to the Variance of an entire population, where $\mu$ is the population mean. When estimating the Variance or std from a sample, the correct formula is *unbiased* is:
$$\hat{Var} = \hat\sigma^2 =\frac{\sum (x_i - \bar{x}) ^ 2}{N-1}$$

```python
#mu, sigma = 0, 0.1 # mean and standard deviation
#s = np.random.normal(mu, sigma, 1000)
```

```python
N = 100
plt.scatter(x=np.random.normal(loc=0, scale = 0.5, size = N), y = np.ones(N), c="black", label="Low variance")
plt.scatter(x=np.random.normal(0, 2, N), y = np.zeros(N), c="y", label="High variance")
```

Note that the variance and standard deviations are also very sensitive to outliers:

```python
print(f'Salary Std in Spain without Messi: {np.round(np.std(mock_salaries), 2)}')
print(f'Salary Std in Spain with Messi: {np.round(np.std(salaries_with_messi), 2)}')
```

```python
plt.hist(np.random.normal(0, 0.5, N), 
         color='b',
         label='low variance',
         alpha=0.2)

plt.hist(np.random.normal(0, 2, N), 
         color='orange',
         label='high variance',
         alpha=0.2)
plt.legend()
plt.show()
```

### Asymmetry (Skewness)

[Wikipedia](https://en.wikipedia.org/wiki/Skewness) In probability theory and statistics, skewness is a measure of the skewness of the probability distribution of a real-valued random variable with respect to his mean. The skewness value can be positive, zero, negative, or undefined.




The relative position of the `mean`, the `median` and the `mode` of the distribution gives us an idea of ​​the `skewness` of our data



The sample skewness is typically calculated as the Fisher-Pearson skewness coefficient. This is the case with the SciPy stats `skew` function. What this coefficient really measures is the difference between the mean and the mode with respect to the dispersion of the set.

```python
plt.hist(df_titanic.age, bins=20)
plt.axvline(df_titanic.age.mean(),
            c="red",
            linewidth= 3.,
            linestyle='--',
            label='mean age') # plotting the mean
   
plt.axvline(df_titanic.age.median(),
            c="orange",
            linewidth= 3.,
            linestyle='--',
            label='median age') # plotting the median

plt.axvline(df_titanic.age.mode()[0],
            c="yellow",
            linewidth= 3.,
            linestyle='--',
            label='modal age') # plotting the mode

plt.title('Histogram of Ages in the Titanic DataFrame', size=20)
plt.xlabel('age')
plt.ylabel('frequency')
plt.legend()
plt.show()
```

### Outliers / Outliers

$$\big[ Q_1 - 1.5 * (Q_3 - Q_1 ), Q_3 + 1.5 * (Q_3 - Q_1 ) \big]$$

This is the standard method of deciding if you are an outlier, although outliers are always going to be a problem.

The $(Q_3 - Q_1 ) := IQR$ value is known as the "interquartile range".


**Note:** There are other ways to mark observations as outliers, and the $1,5$ value can be changed to any non-negative $k$.
Depending on the distribution you have, you can calculate it asymmetrically.

```python
#Max

def give_me_the_outliers (series):
    
    outliers = []
    
    # 1. Range
        
    q1 = series.quantile(0.25)
    q3 = series.quantile(0.75)
    
    iqr = q3-q1
    
    lower_fence = q1 - iqr*1.5
    upper_fence = q3 + iqr*1.5
    
    # 2. Values outside of that
    
    for element in series.values:
        if element > upper_fence or element < lower_fence:
            outliers.append(element)
            
    print(f"Lower fence: {lower_fence}")
    print(f"Upper fence: {upper_fence}")

    
    return outliers
```

### Covariance

[Wikipedia:](https://en.wikipedia.org/wiki/Covariance)

In probability theory and statistics, covariance is a measure of the joint variability of two random variables. If the largest values ​​of one variable correspond mainly to the largest values ​​of the other variable, and the same occurs with the values smaller, (that is, the variables tend to show a similar behavior), the covariance is positive In the opposite case, when the larger values ​​of one variable correspond mainly to the smaller values ​​of the other, (that is, variables tend to show the opposite behavior), the covariance is negative.
**The sign of the covariance shows, therefore, the trend in the linear relationship between the variables**.







The formula for the covariance between two variables ($j$, and $k$) is similar to that for the variance:

$$q_{jk}=\frac{1}{N-1}\sum_{i=1}^{N}\left(  x_{ij}-\bar{x}_j \right)  \left( x_{ik}-\bar{x}_k \right)$$

**Note:** can you see that if $j=k$ the covariance is just the variance?

### Pearson Correlation

The magnitude of the covariance is not easy to interpret because it is not normalized and therefore depends on the magnitudes of the variables. However, the normalized version of the covariance, the Pearson correlation coefficient, shows by its magnitude the strength of the linear relationship. The formula is

$$\rho_{X,Y} = \frac{Cov(X,Y)}{\sigma_X \sigma_Y}$$


where $\sigma_X$ and $\sigma_Y$ are the respective standard deviations.

Pearson's correlation coefficient is a statistic that measures the linear correlation between two variables X and Y. It has a value between +1 and -1. A value of +1 is full positive linear correlation, 0 is no linear correlation, and -1 is full negative linear correlation.


**WARNING!** The fact that the correlation (or covariance) is zero does not mean that there are no patterns

An alternative correlation coefficient known as [Spearman's correlation coefficient](https://en.wikipedia.org/wiki/Spearman%27s_rank_correlation_coefficient) solves this problem by computing Pearson's correlation for the ranks . The sign of the Spearman correlation indicates the direction of the association between X (the independent variable) and Y (the dependent variable). If Y tends to increase as X increases, Spearman's correlation coefficient is positive. If Y tends to decrease as X increases, Spearman's correlation coefficient is negative. A Spearman correlation of zero indicates that there is no tendency for Y to increase or decrease as X increases. Spearman's correlation increases in magnitude as X and Y get closer to being perfectly monotone functions of each other. When X and Y are perfectly monotonically related, the Spearman correlation coefficient becomes 1.

For more information, see [this post](https://towardsdatascience.com/clearly-explained-pearson-v-s-spearman-correlation-coefficient-ada2f473b8) in towardsdatascience.
