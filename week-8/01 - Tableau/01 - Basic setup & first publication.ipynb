# Importing

- Importing

- Let's import `co2.csv`.

- You get a sample of 100 rows by default, but you can tune that number:

#### Make sure the data types are correct

- Click on the hastagh on the "Año" column and change it to `Date & Time`. The rest should be in `String` and `Number (decimal)`

# Exploration

- Adjust the view so you don't miss any info or have to scroll sideways

- Get hue by country: Drag `Nombre del país` to the **Marks** box. Then, drag it to colors.

- Filter: SUM of C02 > 5 only: drag from the left panel `Nombre del país`. Click on top and then on By field. Then input 5.

- Let's filter one country out: On the filter box, click on `Nombre del país` and Show filter. Uncheck Russia box.

- Let's include co2 per capita: from the Measures menu, drag it into the Rows box in the upper section

- Convert the upper one into an area chart: Mark box > dropdown menu > area

- Drag `Nombre del país` into `Label` in the Marks box

- Click on Sheet 1, on the upper section, and change the name to `Co2 in kt and Co2in tons`

### Publish your visualization

- On the upper menu, click on save. Log in or sign up in the pop-window
