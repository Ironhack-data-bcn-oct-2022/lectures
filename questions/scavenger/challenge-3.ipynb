```python
import requests 
import json
import os
from dotenv import load_dotenv
import pandas as pd
from pandas import json_normalize
import time
import datetime as DT

import os
from dotenv import load_dotenv
load_dotenv()
```

```python
bootcamp_url = "https://api.github.com/search/code?q=filename:.scavengerhunt+repo:ironhack-datalabs/scavenger+sort:filename"

file_url="https://api.github.com/repositories/162375181/contents/98750/.0001.scavengerhunt?ref=9308ccc8a4c34c5e3a991ee815222a9691c32476"

repo="ironhack-datalabs/scavenger"
def requestGithub(url):
    api_key = os.getenv("gh_api_key")
    if not api_key:
        raise ValueError( "No hay token.")
    print(f"Estoy haciendo request a {url}")
    parameters  = {"Authorization" : f"token {api_key}"}
    response = requests.get (url = url, params = parameters).json()

    return response
```

```python
test = requestGithub(bootcamp_url)
list_of_dict = (test['items'])
list_of_dict
```

```python
file_url = {}
for dict_ in list_of_dict:
    file_url[dict_['name']] = dict_['url']
file_text_sorted = dict(sorted(file_url.items()))
file_text_sorted
```

```python
downloads = list()
for url in file_text_sorted.values():
    text = requestGithub(url)
    downloads.append(text['download_url'])
downloads
```

```python

api_key = os.getenv("gh_api_key")
parameters  = {"Authorization" : f"token {api_key}"}
for url in downloads:
    text = requests.get (url =  url, params = parameters).text
    print(text)
```
