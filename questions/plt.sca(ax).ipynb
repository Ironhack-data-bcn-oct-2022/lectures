# Sca

### [Docs](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.sca.html): 

`matplotlib.pyplot.sca(ax)`

Set the current Axes to ax and the current Figure to the parent of ax.

### Libraries & data

```python
import matplotlib.pyplot as plt
from scipy import sin, cos

# One X, two Ys
x = [1,2,3,4,5,6,7,8,9]
y1 = sin(x)
y2 = cos(x)
y3 = sin(x) * 2
```

# One row & one column

- What I'm plotting: two Y's in the same X

```python
plt.plot(x, y1, y2)
plt.show()
```

# Two rows & one column

- What I'm plotting: two Y's in the same X

```python
f, ax = plt.subplots(2,1) #(2, 1) -> two rows, one column

plt.plot(x,y1) #x, sin(x)
plt.plot(x,y2) #x, cos(x)
plt.show()
```

## Two rows, one column

- What I'm plotting: two Y's relating to the same X but separated

```python
#subplots(2, 1) -> 2 rows for X axis (one on top of the other) & 1 row for Y axis: one column
f, ax = plt.subplots(2,1)


plt.sca(ax[0]) 
plt.plot(x,y1) #x, sin(x)


plt.sca(ax[1]) 
plt.plot(x,y2) #x, cos(x)
plt.show()
```

## One row, two columns

- What I'm plotting: two Y's in the same X

```python
#subplots(1, 2) -> 1 row for X axis & 2 rows for Y axis: two columns side by side
f, ax = plt.subplots(1, 2)


plt.sca(ax[0])
plt.plot(x,y1) #x, sin(x)


plt.sca(ax[1]) 
plt.plot(x,y2) #x, cos(x)

plt.show()
```

- What I'm plotting: four different Y's that share the same X

```python
#subplots(1, 4) -> 1 row for X axis & 4 rows for Y axis: two columns side by side
f, ax = plt.subplots(1,4)

plt.sca(ax[0]) 
plt.plot(x,y1) #x & sin(x)


plt.sca(ax[1])
plt.plot(x,y2) #x & cos(x)


plt.sca(ax[2])
plt.plot(x,y3) #x & sin(x) * 2


plt.sca(ax[3])
plt.plot(x,y3+y1) #cos(x) + sin(x)
plt.show()
```

- What I'm plotting: four different Y's that share the same X

```python
#subplots(4, 1) -> 4 rows for X axis & 1 rows for Y axis: 4 rows on top of each other
f, ax = plt.subplots(4,1)

plt.sca(ax[0]) 
plt.plot(x,y1) #x & sin(x)


plt.sca(ax[1])
plt.plot(x,y2) #x & cos(x)


plt.sca(ax[2])
plt.plot(x,y3) #x & sin(x) * 2


plt.sca(ax[3])
plt.plot(x,y3+y1) #x & cos(x) + sin(x)
plt.show()
```

```python
import pandas as pd
fitbit = pd.read_csv("data/Fitbit2.csv")
```

### Using plt.sca(): Set the current Axes to ax

```python
fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(12,8))

#PLOT 1

#My data
step = fitbit['Steps'].groupby(fitbit['Months_encoded']).median()

#The axes where I'll plot it
plt.sca(axs[0]) #Set the current Axes to ax 
step.plot(kind='line')

#Details
plt.title("Median Steps by Month")
plt.xlabel('Months')
plt.ylabel('Steps')

#PLOT 2

#My data
wsteps = fitbit['Steps'].groupby(fitbit['Work_or_Weekend']).median()

#The axes where I'll plot it
plt.sca(axs[1]) #Set the current Axes to ax 
wsteps.plot(kind='bar')

#Details
plt.title("Median Steps by Work or Weekend")
plt.xlabel('Work or Weekend')
```

### Not using plt.sca(), but setting the axis when plotting

```python
fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(12,8))

#PLOT 1

#My data
step = fitbit['Steps'].groupby(fitbit['Months_encoded']).median()

#The axes where I'll plot it
step.plot(kind='line', ax = axs[1]) #HERE: it would meet the same function as plt.sca()

#Details
plt.title("Median Steps by Month")
plt.xlabel('Months')
plt.ylabel('Steps')

#PLOT 2

#My data
wsteps = fitbit['Steps'].groupby(fitbit['Work_or_Weekend']).median()

#The axes where I'll plot it
wsteps.plot(kind='bar', ax = axs[0]) #HERE: it would meet the same function as plt.sca()

#Details
plt.title("Median Steps by Work or Weekend")
plt.xlabel('Work or Weekend')
```
