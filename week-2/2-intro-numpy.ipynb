## What is Numpy?

```python
import numpy as np
```

NumPy is a library for the Python programming language that provides support for creating large multidimensional arrays and vectors, along with a large collection of high-level mathematical functions to operate on them. NumPy's forerunner, Numeric, was originally created by Jim Hugunin with contributions from several other developers. In 2005, Travis Oliphant created NumPy by incorporating features from competing Numarray into Numeric, with extensive modifications. NumPy is open source software and has many contributors.

The core of the NumPy package is the ndarray object. This encapsulates n-dimensional arrays of homogeneous data types, with many operations being performed in code compiled for performance. There are several important differences between NumPy arrays and standard Python sequences:

## Work with numeric data
"Data" in data analysis often refers to numerical data, for example stock prices, sales figures, sensor measurements, sports scores, database tables, etc. The Numpy library provides specialized data structures, functions, and other tools for numerical computing in Python. Let's work through an example to see why and how to use Numpy to work with numerical data.

Suppose we want to use climate data such as temperature, rainfall, and humidity to determine if a region is suitable for growing apples. A simple approach to do this would be to formulate the relationship between the annual yield of apples (tonnes per hectare) and climatic conditions such as mean temperature (in degrees Fahrenheit), rainfall (in millimeters), and mean relative humidity (in percent) as a linear equation.

`yield_of_apples = w1 *temperature + w2* rainfall + w3 * humidity`

We are expressing apple yield as a weighted sum of temperature, rainfall, and humidity. This equation is an approximation, as the actual relationship may not be linear and other factors may be involved. But a simple linear model like this usually works well in practice.

Based on a statistical analysis of historical data, we can obtain reasonable values ​​for the weights w1, w2 and w3. Here is an example of values:

```python
w1, w2, w3 = 0.3, 0.2, 0.5
```

We can now substitute these variables into the linear equation to predict the yield of the apples.

```python
temp = 73
rainfall = 67
humidity = 43
```

`yield_of_apples = w1 *temperature + w2* rainfall + w3 * humidity`

To make the above calculation a bit easier for multiple regions, we can represent the climate data for each region as a vector, i.e. a list of numbers.

```python
w1, w2, w3 = 0.3, 0.2, 0.5

temp = 73
rainfall = 67
humidity = 43
```

The three numbers in each vector represent the temperature, precipitation, and humidity data, respectively.
We can also represent the set of weights used in the formula as a vector.

```python
weights_as_lists = [0.3, 0.2, 0.5]
kanto_region = [73, 67, 43]
```

How can we calculate the cultivation of each region?

```python
weights_as_lists * kanto_region
```

```python
def calculation (wei, region):
    counter = 0

    for j, k in zip (wei, region):
        counter = counter + j*k
    return counter
```

The * operator performs an elementwise multiplication of two arrays if they are the same size. The sum method calculates the sum of the numbers in an array.

## Advantages of using Numpy arrays

Numpy arrays offer the following advantages over Python lists for operating on numeric data:

- Ease of use: You can write small, concise and intuitive mathematical expressions like `(kanto * weights).sum()` instead of using loops and custom functions like crop().
- Performance: Numpy's operations and functions are implemented internally in C++, which makes them much faster than using Python statements and loops that are interpreted at runtime

### DIAGONAL Array
Returns a two-dimensional array with ones on the diagonal and zeros on the rest. The syntax is:
`np.eye(N, M=None, k=0, dtype=<class 'float'>, order='C')`
Where:
- **N**: INT Specifies the number of lines in the generated array.
- **M**: INT (Optional) The default value is None. Specifies the number of columns in the array to be generated. When it is None, it is equal to N.
- **k**: INT (Optional) The default value is 0. 0 is the main position of the diagonal. If k is positive, create the upper diagonal with the offset of k; if k is negative, create the lower diagonal with the offset of k.
- **dtype**: data type (Optional) The default value is float. Specifies the data type of the created array element.

```python
np.eye(5,5, dtype=int) #dtype: float / int
```

```python
another_array = np.eye(5,5, k=2, ) # k is the main position of the diagonal. I
another_array
```
